# Assignment

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

# 1. Count of one categorical feature

```python
df['Foundation']
```

```python
sns.countplot(data=df, x='Foundation')
```

```python
## ingin mengurutkan nilainya
foundation_count = df.groupby(['Foundation'],as_index=False).size()
foundation_count = foundation_count.sort_values(['size'],ascending=False)
```

```python
foundation_count
```

```python
sns.barplot(x=foundation_count['Foundation'],
            y=foundation_count['size'])
```

# 2. Distribution of one continuous feature

```python
df['MSSubClass']
```

```python
sns.displot(x=df['MSSubClass'], kde=True, color='Red')

# Add labels and title to the plot
plt.xlabel('The Building Class')
plt.ylabel('Count')
plt.title('Distribution of The Building Class')
```

# 3. Categorical-continuous

```python
df[['Functional','SalePrice']]
```

```python
sns.boxplot(x=df['Functional'],
            y=df['SalePrice'])

## labels and title to the plot
plt.xlabel('Home Functionality Rating')
plt.ylabel('Sale Price')
plt.title('Relationship Between Home Functionality Rating and Sale Price of The Property')
```

# 4. Continuous-continuous

```python
## Cara 1
sns.scatterplot(x=df['YearBuilt'],y=df['SalePrice'])
```

```python
## cara 2
sns.lmplot(x='YearBuilt', y='SalePrice', data=df, scatter_kws={'color': 'Red'}, line_kws={'color': 'green'})

## labels and title to the plot
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.title('Relationship Between Year Built and The Sale Price of the Property')
```
